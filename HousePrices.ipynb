```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
dados = pd.read_csv('train.csv')
```

```python
dados.head()
```

```python
dados.shape
```

```python
dados.info()
```

```python
y = dados['SalePrice']
dados.drop(columns='SalePrice', inplace=True)
dados.head()
```

```python
dados.dtypes
```

```python
colunas_int = dados.select_dtypes(include=['int64']).columns[:-1]
colunas_obj = dados.select_dtypes(include=['object']).columns
colunas_float = dados.select_dtypes(include=['float64']).columns
```

```python
colunas_float
```

```python
X = dados[colunas_int.append(colunas_float)]
```

```python
from category_encoders.ordinal import OrdinalEncoder
enc = OrdinalEncoder(cols=colunas_obj)
enc.fit(dados)
dados_ord = enc.transform(dados)
```

```python
dados_ord.head()
```

```python
X = pd.concat([X, dados_ord], axis = 1)
```

```python
X.shape
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python
X.fillna(-1, inplace=True)
X.isnull().sum()
```

```python
X.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()
modelo.fit(X_train, y_train)
predicao = modelo.predict(X_test)

mean_squared_error(predicao, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

modelo = RandomForestRegressor()
modelo.fit(X_train, y_train)
predicao = modelo.predict(X_test)

mean_squared_error(predicao, y_test)
```

**Testando o modelo**

```python
dt_test = pd.read_csv('test.csv')
dt_test.head()
```

```python
dt_test.fillna(-1, inplace=True)
```

```python
enc = OrdinalEncoder(cols=colunas_obj)
enc.fit(dt_test)
dados_ord = enc.transform(dt_test)
```

```python
X = dt_test[colunas_int.append(colunas_float)]
```

```python
X = pd.concat([X, dados_ord], axis = 1)
```

```python
X.columns
```

```python
previsao = modelo.predict(X)
```

```python
sub = pd.Series(previsao, index=dt_test['Id'], name='SalePrice')
sub.to_csv('sample_submission.csv', header = True)
```

**Usando seletor de variaveis**

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression, f_classif
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

k=10
for k in range(2,X_train.shape[1],2):

    #Instanciando o modelo de selecao de features
    selector = SelectKBest(score_func=f_regression, k=k)

    #Treinando o seletor
    selector.fit(X_train, y_train)
    X_train2 = selector.transform(X_train)

    #Aplicando nos dados de testes
    X_test2 = selector.transform(X_test)

    modelo = RandomForestRegressor(n_estimators=1000, n_jobs=-1, random_state=0)
    modelo.fit(X_train2, y_train)
    predicao = modelo.predict(X_test2)
    
    score = mean_squared_error(y_test, predicao)
    print('k = {} - MAE = {}'.format(k, score))

k_vs_score.append(score)
```

```python
selector = SelectKBest(score_func=f_regression, k=50)
#Treinando o seletor
selector.fit(X_train, y_train)
X_train2 = selector.transform(X_train)

#Aplicando nos dados de testes
X_test2 = selector.transform(X_test)

modelo = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=0)
modelo.fit(X_train2, y_train)
predicao = modelo.predict(X_test2)

mask = selector.get_support()
colunas = X_test.columns[mask]

score = mean_squared_error(y_test, predicao)
print('k = {} - MAE = {}'.format(k, score))
```
